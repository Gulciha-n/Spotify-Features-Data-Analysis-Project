```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks = pd.read_csv("/content/tracks.csv")
df_tracks.head()
```

```python
#total number of null values
pd.isnull(df_tracks).sum()
```

```python
df_tracks.info()
```

```python
sorted_df = df_tracks.sort_values("popularity", ascending=True).head(10)
sorted_df
```

```python
# describe() : statistics columns (count,mean,std,main,%25,%50,%75,max)
# transpose() : rows and columns are swapped
df_tracks.describe().transpose()
```

```python
most_popular = df_tracks.query("popularity>90" , inplace=False).sort_values("popularity",ascending=False)
most_popular[:10]
```

```python
df_tracks.head()
```

```python
df_tracks.set_index('release_date')
df_tracks['release_date'] = pd.to_datetime(df_tracks['release_date'],format='ISO8601')
df_tracks.set_index('release_date', inplace=True)
df_tracks.head()
# set_index : The release date column was set as the index
```

```python
df_tracks[["artists"]].iloc[18]
# iloc (index location) : access to data
```

```python
# converting milisecond to second
df_tracks["duration"] = df_tracks["duration_ms"].apply(lambda x : round(x/100))
df_tracks.drop("duration_ms",inplace=True,axis=1)
```

```python
df_tracks.duration.head()
```

```python
# heatmap
numeric_df = df_tracks.drop(["key","mode","explicit"], axis=1).select_dtypes(include=['float64', 'int64'])
corr_df = numeric_df.corr(method='pearson')
heatmap = sns.heatmap(corr_df,annot=True , fmt=".1g",vmin=-1,vmax=1,center=0,cmap="inferno",linewidths=1,linecolor="Black")
heatmap.set_title("Correlation HeatMap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)

# +1: Full positive linear relationship
#  0: No relationship
# -1: Full negative linear relationship
# vmin : minimum value , cmap : color map , annot : annotate , fmt : format (.1g : decimal)
```

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
# This code randomly samples 0.4% of the rows from the `df_tracks` DataFrame and prints the number of rows in the sample.
```

```python
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df , y="loudness",x="energy",color="b").set_title("Loudness vs Energy Correlation")
# This code sets up a figure with specific dimensions, and then creates a regression plot using the sample_df DataFrame,
# examining the relationship between the "loudness" and "energy" features, with the plot titled "Loudness vs Energy Correlation.
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df , y="popularity", x="acousticness", color="r").set_title("Popularity vs Acousticness Correlation")
```

```python
df_tracks["dates"] = df_tracks.index.get_level_values("release_date")
df_tracks.dates = pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
```

```python
sns.displot(years,discrete=True,aspect=2,height=5,kind="hist").set(title= "Number of songs per year")

# visualizes the number of songs per year as a histogram
# years: The source of the data for the histogram.
# discrete=True: Specifies that the values are discrete, such as years.
# aspect=2: Sets the aspect ratio, making the plot twice as wide.
# height=5: Sets the height of the plot to 5 inches.
# kind="hist": Specifies that the plot type is a histogram.
```

```python
total_dr = df_tracks.duration
fig_dims = (18,7)
fig , ax = plt.subplots(figsize = fig_dims)
fig  = sns.barplot(x = years , y = total_dr , ax = ax , err_kws={'linewidth': False}).set(title="Years vs Duration")
plt.xticks(rotation=90)
plt.show()

# err_kws={'linewidth': False}: Sets the width of error bars to zero.
# sns.barplot(): Creates a bar plot.
```

```python
total_dr = df_tracks.duration
sns.set_style(style="whitegrid")
fig_dims = (10,5)
fig , ax = plt.subplots(figsize = fig_dims)
fig  = sns.lineplot(x =years , y=total_dr , ax=ax).set(title="Years vs Duration")
plt.xticks(rotation=60)
plt.show()
```

```python
sns.set_style(style="darkgrid")
plt.figure(figsize=(10,5))
famous = df_tracks.sort_values("popularity",ascending=False).head(10)
sns.barplot(y="artists", x = "popularity",data=famous).set(title="Top 5 artist by popularity")
plt.show()
```
